```python
from matplotlib import pyplot as plt

movies = ["Annie Hall", "Ben - Hur", "Casablanca", "Gandhi", "west Side Story"]

numeros_oscars = [5, 11, 3, 8, 10]

xs = [i + 0.1 for i, _ in enumerate(movies)]

plt.bar(xs, numeros_oscars)

plt.ylabel("# de Premiação")

plt.title("Filmes ao Oscar")

plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies )

plt.show()
```

```python
from matplotlib import pyplot as plt

variance = [1,2,4,8,16,32,64,128,256]

bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total = [x + y for x, y in zip(variance, bias_squared)]

xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label = 'variance')
plt.plot(xs, bias_squared, 'r-.', label= 'bias ^2')
plt.plot(xs, total, 'b:', label = 'total error')

plt.legend(loc = 9)
plt.xlabel("complexidade")
plt.title("compromisso entre Polariza")
plt.show()
```

```python
import numpy
import scipy.stats

dados = numpy.random.normal(loc=0.0, scale=1.0, size=10000)

obliquidade = scipy.stats.skew(dados)
print(obliquidade)
```

```python
import numpy
import scipy.stats

dados = numpy.random.normal(loc=0.0, scale=1.0, size=10000)

curtose = scipy.stats.kurtosis(dados)
print(curtose)
```
